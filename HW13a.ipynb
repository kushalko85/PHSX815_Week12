```python
#!bin/python
import numpy as np
import matplotlib.pyplot as plt

```

```python
N_meas = 50
N_exp = 100
ave = []

for i in range(0,N_exp):
    sum =0.
    for j in range(0,N_meas):
        x = np.random.exponential(0.2)
        sum = sum +x
        #print (sum)
    avg = float(sum)/float(N_meas)
    ave.append(avg)
#meas_value_l.append(float(meas_value))
#true_value_l.append(float(true_value))


#Plotting stuffs
plt.figure()
#plt.scatter(true_value_l,meas_value_l)
plt.hist(ave,bins = 30)
plt.xlabel('Average')
plt.ylabel('relative frequency')
plt.title('CLT ddemonstration from exponential distribution')
plt.show()
```
